##A* repeatedly does this
### 1. Pick the best state
### 2. generate all possible next
### 3. calculate their cost
### 4. store them
### 5. Repeat until the goal is found


```python
from heapq import heappush,heappop
```

```python
goal = (
    (1,2,3),
    (4,5,6),
    (7,8,0)
)

start = (
    (0,1,3),
    (4,2,6),
    (7,5,8)
)
```

```python
def misplaced_tiles(state):
    count=0
    for i in range(3):
        for j in range(3):
            if state[i][j]!=0 and state[i][j] != goal[i][j]:
                count=count+1
    return count

print(misplaced_tiles(start))
```

```python
def find_blank(state):
    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i,j
```

```python
def get_neighbors(state):
    x,y=find_blank(state)
    moves=[(-1,0),(1,0),(0,-1),(0,1)]
    neighbors=[]
    for dx,dy in moves:
        nx=x+dx
        ny=y+dy
        if 0 <= nx < 3 and 0 <= ny <3:
            board=[list(row) for row in state]
            board[x][y],board[nx][ny]=board[nx][ny],board[x][y]
            neighbors.append(tuple(map(tuple,board)))

    return neighbors

# neighbors = get_neighbors(start)
# print(neighbors)
```

```python
def reconstruct(parent,current):
    path = []
    while current is not None:
        path.append(current)
        current = parent[current]
    return path[::-1]
```

```python
def astar(start,heuristic):
    pq=[]
    heappush(pq, (heuristic(start),0, start))
    parent={start:None}
    g_cost={start:0}
    expanded=0
    while pq:
        f,g,current = heappop(pq)
        expanded = expanded + 1
        if current == goal:
            return reconstruct(parent,current),expanded
        for neighbor in get_neighbors(current):
            new_g = g + 1
            if neighbor not in g_cost or new_g < g_cost[neighbor]:
                g_cost[neighbor]=new_g
                f_cost=new_g+heuristic(neighbor)

                heappush(pq,(f_cost,new_g,neighbor))
                parent[neighbor]=current

    return None,expanded
```

```python
print("using h1: Misplaced Tiles")
path,expanded=astar(start,misplaced_tiles)
if path is not None:
    for step in path:
        for row in step:
            print(row)
        print()

print('Moves = ',len(path)-1)
print('Nodes Expanded =',expanded)
```
