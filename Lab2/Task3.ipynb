##Task 3: Compare the Heuristics by runnig the A* algorithm using: 
###1. h1: Misplaced Tiles 
###2. h2: Manhattan Distance 
###3. h1 + h2: Combined Heuristic

```python
from heapq import heappush, heappop

# Goal State
goal = (
    (1, 2, 3),
    (4, 5, 6),
    (7, 8, 0)
)

# Start State
start = (
    (5, 8, 2),
    (1, 0, 3),
    (4, 7, 6)
)
```

```python
def misplaced_tiles(state):
    count = 0

    for i in range(3):
        for j in range(3):
            if state[i][j] != 0 and state[i][j] != goal[i][j]:
                count += 1

    return count
```

```python
goal_pos = {}

for i in range(3):
    for j in range(3):
        goal_pos[goal[i][j]] = (i, j)
```

```python
def manhattan_distance(state):

    distance = 0

    for i in range(3):
        for j in range(3):

            tile = state[i][j]

            if tile != 0:
                goal_i, goal_j = goal_pos[tile]
                distance += abs(i-goal_i) + abs(j-goal_j)

    return distance
```

```python
def combined_heuristic(state):
    return misplaced_tiles(state) + manhattan_distance(state)
```

```python
def find_blank(state):

    for i in range(3):
        for j in range(3):
            if state[i][j] == 0:
                return i, j
```

```python
def get_neighbors(state):

    x, y = find_blank(state)

    moves = [
        (-1,0),
        (1,0),
        (0,-1),
        (0,1)
    ]

    neighbors = []

    for dx, dy in moves:

        nx = x + dx
        ny = y + dy

        if 0 <= nx < 3 and 0 <= ny < 3:

            board = [list(row) for row in state]

            board[x][y], board[nx][ny] = board[nx][ny], board[x][y]

            neighbors.append(tuple(map(tuple, board)))

    return neighbors
```

```python
def reconstruct(parent, current):

    path = []

    while current is not None:
        path.append(current)
        current = parent[current]

    return path[::-1]
```

```python
def astar(start, heuristic):

    pq = []

    heappush(pq, (heuristic(start), 0, start))

    parent = {start: None}
    g_cost = {start: 0}

    expanded = 0

    while pq:

        f, g, current = heappop(pq)

        expanded += 1

        if current == goal:
            return reconstruct(parent, current), expanded

        for neighbor in get_neighbors(current):

            new_g = g + 1

            if neighbor not in g_cost or new_g < g_cost[neighbor]:

                g_cost[neighbor] = new_g

                f_cost = new_g + heuristic(neighbor)

                heappush(pq, (f_cost, new_g, neighbor))

                parent[neighbor] = current

    return None, expanded
```

```python
def run_test(name, heuristic):

    print(name)


    path, expanded = astar(start, heuristic)

    if path:

        print("Solution Path:\n")

        for step in path:
            for row in step:
                print(row)
            print()

        print("Total Moves =", len(path)-1)
        print("Nodes Expanded =", expanded)

        return len(path)-1, expanded

    else:
        print("No Solution Found.")
        return None, None
```

```python
moves1, nodes1 = run_test("Heuristic 1: Misplaced Tiles", misplaced_tiles)

moves2, nodes2 = run_test("Heuristic 2: Manhattan Distance", manhattan_distance)

moves3, nodes3 = run_test("Heuristic 3: Combined (h1 + h2)", combined_heuristic)
```
