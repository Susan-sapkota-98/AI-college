##Task 3: Robot Navigation Problem
Problem Statement
A robot is placed in a 5 ×5 grid and must move from the start position (S) to the goal position (G) while
avoiding obstacles (X).

Tasks
1. Represent the grid as a search space.
2. Define:
• State representation
• Initial state
• Goal state
• Valid actions
3. Write a Python program to find a path from the start cell to the goal cell using BFS.
4. Print the shortest path found by the algorithm.
Valid Actions
• Move Up
• Move Down
• Move Left
• Move Right
Expected Output
The program should display:
• The sequence of coordinates visited
