#Task 1: Goat, Wolf and Cabbage Problem
##Problem Statement
A farmer has a goat, a wolf and a cabbage on the west side of a river. He wants to transport all three to
the east side using a boat that can carry only the farmer and one additional item. The following
constraints must always be satisfied:
• The wolf cannot be left alone with the goat.
• The goat cannot be left alone with the cabbage.
##Tasks
1. Formulate the problem as a state-space search problem.
2. Define:
• State representation
• Initial state
• Goal state
• Valid operators/actions
##3. Write a Python program to solve the problem using a search algorithm.


```python
from collections import deque
# for (Farmer, Wolf, Goat, Cabbage)
initial_state=('W','W','W','W')
goal_state=('E','E','E','E')
```

```python
def is_valid(state):
    f,w,g,c=state
    # If wolf and goat are on the same side, but farmer is on the other side
    if w==g and f != w:
        return False
    # If goat and cabbage are on the same side, but farmer is on the other side
    if g==c and f != g:
        return False

    return True

```

```python
def next_state(state):
    f,w,g,c = state
    next_states = []

    # The farmer can move alone (None) or take one item that is on the same side

    moves = [None]
    
    if w == f:
        moves.append('wolf')

    if g == f:
        moves.append('goat')

    if c == f:
        moves.append('cabbage')

    new_f = 'E' if f == 'W' else 'W'

    for item in moves:
        new_w,new_g,new_c = w,g,c

        if item == 'wolf':
            new_w = new_f

        elif item == 'goat':
            new_g = new_f

        elif item == 'cabbage':
            new_c = new_f

        new_state = (new_f,new_w,new_g,new_c)

        if is_valid(new_state):
            next_states.append((new_state,item))

    return next_states
```

```python
from collections import deque

def solve_puzzle():
    queue = deque([(initial_state, [])])
    visited = {initial_state}

    while queue:
        current_state, path = queue.popleft()

        if current_state == goal_state:
            return path

        for state, item_moved in next_state(current_state):

            if state not in visited:
                visited.add(state)

                move_description = (
                    f"Farmer takes {item_moved}"
                    if item_moved
                    else "Farmer goes alone"
                )

                new_path = path + [(move_description, state)]

                queue.append((state, new_path))

    return None
```

```python
solution = solve_puzzle()

if solution:
    print("Success! Here are the steps to cross the river:\n")
    print(f"Start: All on the West side.")
    print("-" * 40)
    for i, (move, state) in enumerate(solution, 1):
        print(f"Step {i}: {move}")
        # Formatting the output nicely
        west_side = [name for name, side in zip(['Farmer', 'Wolf', 'Goat', 'Cabbage'], state) if side == 'W']
        east_side = [name for name, side in zip(['Farmer', 'Wolf', 'Goat', 'Cabbage'], state) if side == 'E']
        print(f"   [west side]: {', '.join(west_side) if west_side else 'Empty'}")
        print(f"   [east side]: {', '.join(east_side) if east_side else 'Empty'}\n")
else:
    print("No solution found.")
```
