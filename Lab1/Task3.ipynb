##Task 3: Robot Navigation Problem
Problem Statement
A robot is placed in a 5 ×5 grid and must move from the start position (S) to the goal position (G) while
avoiding obstacles (X).

Tasks
1. Represent the grid as a search space.
2. Define:
• State representation
• Initial state
• Goal state
• Valid actions
3. Write a Python program to find a path from the start cell to the goal cell using BFS.
4. Print the shortest path found by the algorithm.
Valid Actions
• Move Up
• Move Down
• Move Left
• Move Right
Expected Output
The program should display:
• The sequence of coordinates visited

```python
from collections import deque
```

```python
grid = [
    ['S', '0', '0', 'X', '0'],
    ['X', 'X', '0', 'X', '0'],
    ['0', '0', '0', '0', '0'],
    ['0', 'X', 'X', 'X', '0'],
    ['0', '0', '0', 'G', '0']
]
rows=len(grid)
cols=len(grid[0])
start=(0,0)
goal=(4,3)
```

```python
# possible moves
moves=[
    (-1,0),
    (1,0),
    (0,-1),
    (0,1)
    ] # up down left write

```

```python
def bfs(start,goal):
    queue = deque([(start,[start])])
    visited=set()

    while queue:
        current, path = queue.popleft()

        if current == goal:
            return path
        
        if current in visited:
            continue

        visited.add(current)

        for dr, dc in moves:
            nr=current[0]+ dr
            nc=current[1]+dc

            if (0 <= nr < rows and
                0 <= nc < cols and
                grid[nr][nc] != 'X' and
                (nr, nc) not in visited):

                queue.append(((nr, nc), path + [(nr, nc)]))

    return None

```

```python
path=bfs(start,goal)

if path:
    print("shortest path found:")
    for node in path:
        print(node)

    print('Total Steps:', len(path)-1)
else:
    print('No Path Found')
```
