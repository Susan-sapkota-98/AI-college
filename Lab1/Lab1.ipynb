```python
from pprint import pprint
from time import sleep
```

```python
def is_safe(board,x,y,n):
    #Row test
    for row in range(x):
        if board[row][y]=='Q':
            return False

    #Left upper diagonal test

    r=x-1
    c=y-1
    while(r>=0 and c>=0):
        if (board[r][c]=='Q'):
            return False
        r=r-1
        c=c-1
    
    #Right upper diagonal test

    r=x-1
    c=y+1
    while(r>=0 and c<n):
        if (board[r][c]=='Q'):
            return False
        r=r-1
        c=c+1

    return True
        


```

```python

n=int(input('Enter the number of Q'))
board=[['_']*n for i in range(n)]
def nQueensSolver(board,row,n):
    if (row>=n):
        return True
    for col in range(n):
        if is_safe(board,row,col,n):
            board[row][col]='Q'
            if(nQueensSolver(board,row+1,n)):
                return True
            board[row][col]='_'
    return False
if(nQueensSolver(board,0,n)):
    pprint(board)

else:
    pprint("No solution")
```
