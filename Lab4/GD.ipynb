```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(10,30,0.13)
y=2.222*X+3.333
```

```python
split=int(len(X)*0.8)
X_train,y_train=X[:split],y[:split]
x_test,y_test=X[split:],y[split:]
```

```python
def plot_data(X_train, y_train, X_test, y_test, predictions=None):
    plt.scatter(X_train, y_train, c='b', label='Training Data')
    plt.scatter(X_test, y_test, c='r', label='Testing Data')
    if predictions is not None:
        plt.plot(X_test, predictions, c='g', label='Predictions')
        plt.xlabel('X')
        plt.ylabel('y')
        plt.title('X vs y') 
        plt.legend()
        plt.show()

plot_data(X_train,y_train,x_test,y_test)
```

```python
w=np.random.randn()
b=np.random.randn()
```

```python
def loss_func(y_actual,y_pred):
    return np.sum(np.square(y_actual-y_pred))/len(y_actual)
```

```python
y_pred=x_test*w+b
plot_data(X_train,y_train,x_test,y_test,predictions=y_pred)
```

```python
lr=0.1
epochs=100
train_loss_list,test_loss_list=[],[]

for epoch in range(epochs):
# Forward pass
    train_preds = X_train*w + b
    train_loss = loss_func(y_train, train_preds)
# Compute gradients
    gradientW = (-2/len(X_train)) * np.sum((y_train - train_preds)*X_train)
    gradientb = (-2/len(X_train)) * np.sum(y_train - train_preds)
# Update parameters
    w = w - gradientW*lr
    b = b - gradientb*lr
# Evaluate on test data
    test_preds = x_test*w + b
    test_loss = loss_func(y_test, test_preds)
    train_loss_list.append(train_loss)
    test_loss_list.append(test_loss)
```
